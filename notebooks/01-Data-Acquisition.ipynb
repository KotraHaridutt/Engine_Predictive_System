```python
import pandas as pd
import numpy as np
import os
```

```python
column_names = ['unit_number', 'time_cycle', 
                'setting_1', 'setting_2', 'setting_3',
                'sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5',
                'sensor_6', 'sensor_7', 'sensor_8', 'sensor_9', 'sensor_10',
                'sensor_11', 'sensor_12', 'sensor_13', 'sensor_14', 'sensor_15',
                'sensor_16', 'sensor_17', 'sensor_18', 'sensor_19', 'sensor_20',
                'sensor_21']


train_path = '/workspaces/Engine_Predictive_System/data/train_FD001.txt'
test_path = '/workspaces/Engine_Predictive_System/data/test_FD001.txt'
rul_path = '/workspaces/Engine_Predictive_System/data/RUL_FD001.txt'
```

```python
df_train = pd.read_csv(train_path, 
                       sep='\s+', 
                       header=None, 
                       names=column_names)

df_train = df_train.dropna(axis='columns', how='all')

print("--- Training Data (First 5 Rows) ---")
print(df_train.head())
```

```python

df_test = pd.read_csv(test_path, 
                      sep='\s+', 
                      header=None, 
                      names=column_names)

df_test = df_test.dropna(axis='columns', how='all')

print("\n--- Test Data (First 5 Rows) ---")
print(df_test.head())
```

```python
df_rul = pd.read_csv(rul_path, 
                     sep='\s+', 
                     header=None, 
                     names=['RUL']) # Only has one column

df_rul = df_rul.dropna(axis='columns', how='all')

print("\n--- RUL Truth Data (First 5 Rows) ---")
print(df_rul.head())
```
